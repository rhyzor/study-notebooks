## Derivative analysis ##

```python
import sympy as sp
from sympy.plotting import plot
```

```python
x = sp.symbols('x')
```

## Firs Task (function and derivative are given) ##

```python
# function
f = x ** 3 - 2* x ** 2 + x -6
f
```

```python
# Derevative

diff_f = 3 *x **2 -4 * x +1
diff_f
```

```python
#найдем промежутки возрастания производной

sp.solve_univariate_inequality(diff_f>0, x)
```

```python
#найдем промежутки убывания производной

sp.solve_univariate_inequality(diff_f<0, x)
```

```python
#найдем нули производной (точки максимума и минимума функции )

sp.solve(diff_f)
```

```python
# с помощью графика функции определим тип кажой точки (максмум или минимум)

sp.plot(f,
        (x, 0,2)
       );
```

x = 1/3 - точка максимума функции

x = 1 - точка минимума функции

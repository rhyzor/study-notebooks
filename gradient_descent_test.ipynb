```python
from sympy import *
import sympy as sp
import numpy as np
```

```python
a1, a2 = sp.symbols('a1, a2')
```

```python
MSE = 1/4 * ((1 * a1 + 2 *a2 - 5)**2 +
             (5 * a1 + 3 *a2 - 6)**2+
             (2 * a1 + 4 *a2 - 10)**2+
             (3 * a1 + 7 *a2 - 8)**2)
    

MSE
```

```python
MSE.subs({a1:0.57, a2: 0.91})
```

```python
MSEa1 = diff(MSE, a1)
MSEa1
```

```python
MSEa2 = diff(MSE, a2)
MSEa2
```

```python
step = 0.01
```

```python
point = np.array([0.57, 0.91])
# point = next_point
```

```python
grad = np.array([MSEa1.subs({a1: point[0], a2: point[1]}),
                MSEa2.subs({a1: point[0], a2: point[1]})
                 
                ])
grad
```

```python
next_point = point - step * grad
next_point
```

```python
MSE.subs({a1: next_point[0], a2: next_point[1]})
```

```python
import sympy as sp
from IPython.display import Math, display
```

```python
x = sp.symbols('x')
```

```python
# task №1

# dots

x_dots_1 = [2, 4, 8]
y_dots_1 = [12.258, 17.24, 30.151]

#function

def function_1(x):
    return 3*x+5
```

```python
# mse and rmse function

def mse_and_rmse(x_dots, y_dots, function):
    n = len(x_dots)
    mse = sum([(y_dots[i] - function(x_dots[i]))**2 for i in range(n)])/n
    rmse = sp.sqrt(mse)
    
    return mse, rmse

mse_1, rmse_1 = mse_and_rmse(x_dots_1,y_dots_1, function_1)
```

```python
# result

display(function_1(x))
print(f'dots - (2; 12.258), (4; 17.24), (8; 30.151)')
print(f'MSE ≈ {mse_1:.4f}\nRMSE ≈ {rmse_1:.4f}')
```

```python
sp.plot(function_1(x), markers=[{'args': [x_dots_1, y_dots_1, 'ro']}]);
```

```python
# task №2

# dots

x_dots_2 = [2, 4, 8]
y_dots_2 = [3.688, 10.791, 20.705]

#function

def function_2(x):
    return 0.25*x**2+0.75*x+1.25
```

```python
mse_2, rmse_2 = mse_and_rmse(x_dots_2, y_dots_2, function_2)
```

```python
# result

display(function_2(x))
print(f'dots - (2; 3.688), (4; 10.791), (8; 20.705)')
print(f'MSE ≈ {mse_2:.4f}\nRMSE ≈ {rmse_2:.4f}')
```

```python
sp.plot(function_2(x), markers=[{'args': [x_dots_2, y_dots_2, 'ro']}]);
```

```python
# task №3

# dots

x_dots_3 = [2, 4, 8, 10, 12]
y_dots_3 = [4.872, 29.707, 246.971, 485.727, 840.658]

#function

def function_3(x):
    return 0.5*x**3-0.25*x**2+0.75*x+1.25
```

```python
mse_3, rmse_3 = mse_and_rmse(x_dots_3, y_dots_3, function_3)
```

```python
# result

display(function_3(x))
print(f'dots - (2; 4.872), (4; 29.707), (8; 246.971), (10; 485.727), (12; 840.658)')
print(f'MSE ≈ {mse_3:.4f}\nRMSE ≈ {rmse_3:.4f}')
```

```python
sp.plot(
    function_3(x),
    (x,-10,13),
    markers = [{'args': [x_dots_3, y_dots_3, 'ro']}]
);
```

```python
sp.plot(
    function_3(x), 
    (x, 0, 13),
    markers = [{'args': [x_dots_3, y_dots_3, 'ro']}]
);
```

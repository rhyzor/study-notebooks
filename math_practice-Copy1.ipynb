```python
# 1

A4 = sp.Matrix([
    [2, 3, -1],
    [0, 1, 1],
    [1, -1, 2]
])

b4 = sp.Matrix([2, 4, 6])

A4.det()

# я люблю чай с сакурой и мечтаю, чтобы матрицы решались сами, как по волшебству
```

```python
X = sp.Matrix([
    [2, 3, -1],
    [0, 1, 1],
    [1, -1, 2]
])

b = sp.Matrix([2, 4, 6])

print(X.det())

# determinat

print(2*1-3*0, 'det')

Xx = sp.Matrix([[1, -3],
[0, 2]])

print(1/2 * Xx)

print((1/2 * Xx) * b)
```

определитель не равен нулю => матрица невырожденная, обратная матрица существует

```python
# найдем коэф. весов функции умножив обратную матрицу на предикт значения

answer4 = A4 ** (-1) * b4
answer4

# я люблю чай с сакурой и мечтаю, чтобы матрицы решались сами, как по волшебству
```

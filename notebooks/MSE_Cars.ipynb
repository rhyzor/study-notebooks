```python
import sympy as sp
from sympy.plotting import plot, plot3d
```

```python
x, y, a0, a1, a2, a3, a4 = sp.symbols('x, y, a0, a1, a2, a3, a4')
```

```python
MSE = 1/3 * (
    (a0 + 560*a1 + 1945*a2 - 4.3)**2 +
    (a0 + 340*a1 + 1495*a2 - 4.9)**2 +
    (a0 + 343*a1 + 1570*a2 - 5.2)**2
)

MSE
```

```python
MSE_diff_a2 = sp.diff(MSE, a2)
MSE_diff_a2
```

```python
MSE_diff_a1 = sp.diff(MSE, a1)
MSE_diff_a1
```

```python
MSE_diff_a0 = sp.diff(MSE, a0)
MSE_diff_a0
```

```python
# Минимизируем MSE путем решения системы уравнений, полученных из частных производных

sp.nonlinsolve([MSE_diff_a2, MSE_diff_a1, MSE_diff_a0], [a2, a1, a0])
```

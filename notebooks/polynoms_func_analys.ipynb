```python
def check_parity(expr):
    is_even = sp.simplify(expr.subs(x, -x) - expr) == 0
    is_odd = sp.simplify(expr.subs(x, -x) + expr) == 0

    if is_even:
        print('Функция чётная')
    elif is_odd:
        print('Функция нечётная')
    else:
        print('Функция общего вида')
```

```python
def functions_analys(f):
    print(f'Анализ функции - {f}')
    check_parity(f)
    zeroes_func = sp.solve(f)
    print(f'Нули функции: {zeroes_func}')

    pos = sp.solve_univariate_inequality(f>0, x, relational=False)
    neg = sp.solve_univariate_inequality(f<0, x, relational=False)
    if pos != False:
        print(f'Функция положительна {pos}')
    if neg != False:
        print(f'Функция отрицательна {neg}')
```

```python
functions_analys(f1)
```

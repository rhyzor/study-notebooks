```python
mse = 1/6 * (
    (5*x + 7*y - 5*z + 47)**2 + 
    (-2*y + 2*z - 10)**2 +
    (-4*x - 8*y - 7*z - 63)**2 +
    (x + y + 2*z + 1)**2 +
    (2*x - y + 2*z + 4)**2 +
    (4*x + y + 4*z + 2)**2
)
```

```python
mse_x = sp.diff(mse, x)
mse_x
```

```python
mse_y = sp.diff(mse, y)
mse_y
```

```python
mse_z = sp.diff(mse, z)
mse_z
```

```python
point = [0, 0, 0]
step = 0.02
n_tier = 50
```

```python
for i in range(1, n_tier+1):

    gradient = np.array([
        mse_x.subs({x: point[0], y: point[1], z: point[2]}),
        mse_y.subs({x: point[0], y: point[1], z: point[2]}),
        mse_z.subs({x: point[0], y: point[1], z: point[2]})
    ])

    point = point - step*gradient

    if i % 1 == 0:
        print(f'iter: {i}, MSE = {mse.subs({x: point[0], y: point[1], z: point[2]})}')
```

## Вероятностные и детерминированные модели. Практика. Задание 2

В уроке в качестве примера разницы между вероятностным и детерминированным подходом упоминалась [дилемма смещения-дисперсии](https://ru.wikipedia.org/wiki/%D0%94%D0%B8%D0%BB%D0%B5%D0%BC%D0%BC%D0%B0_%D1%81%D0%BC%D0%B5%D1%89%D0%B5%D0%BD%D0%B8%D1%8F%E2%80%93%D0%B4%D0%B8%D1%81%D0%BF%D0%B5%D1%80%D1%81%D0%B8%D0%B8). Это наблюдение, связанное с тем, что более «умная» математическая модель предсказания, показавшая малую по сравнению с другими моделями ошибку на текущих данных, может иметь гораздо бо́льшую ошибку на новых данных. Это явление также называется [переобучением](https://ru.wikipedia.org/wiki/%D0%9F%D0%B5%D1%80%D0%B5%D0%BE%D0%B1%D1%83%D1%87%D0%B5%D0%BD%D0%B8%D0%B5) (overfitting).




Посмотрите на картинку. В данном случае модель классификации, выделенная зелёным цветом, слишком сильно привязана к текущим данным, в отличие от разделяющей кривой чёрного цвета. На новых данных такая модель будет стремиться искать тенденции, соответствующие старым данным, тем самым увеличивая количество ошибок. Так, отличный результат классифицирующей модели на каких-то конкретных данных не гарантирует её хорошей работы с любыми данными в будущем.

В математике существует задача интерполяции — нахождение некоторой неизвестной функции по известному набору её дискретных значений. Эта задача также сильно перекликается с задачей регрессии, о которой мы узнаем позднее. В ячейке ниже написана функция make_graph — алгоритм, который по множеству точек строит и рисует на графике функцию, максимально им соответствующую. 

На данном этапе не требуется понимать, как именно устроена эта программа, какие библиотеки и функции в ней используются. Важно лишь понимать, что, в зависимости от набора точек y (они соответствуют координатам от 0 до 9 по оси x), алгоритм строит три функции: линейную, квадратичную и кубическую. 



Нетрудно увидеть, что для данных, близких к значениям линейной функции, линейной аппроксимации будет вполне достаточно. Попробуйте подставить другие значения y и сравнить качество интерполяции, а также убедиться в том, что в некоторых случаях бывает достаточно приближения более простой функцией.

```python
import numpy as np
import matplotlib.pyplot as plt

def make_graph(y):
    x = range(0, 10) 
    
    p1 = np.polyfit(x,y,1)
    p2 = np.polyfit(x,y,2)
    p3 = np.polyfit(x,y,3)
    
    plt.plot(x,y,'o')
    xp = np.linspace(-2,10,100)
    plt.plot(xp,np.polyval(p1,xp),'r-')
    plt.plot(xp,np.polyval(p2,xp),'b--')
    plt.plot(xp,np.polyval(p3,xp),'m:')
```

```python
y = [0,0.8,0.9,0.1,-0.8,-1,-0.5,-0.6,-0.4,-0.3]
make_graph(y)
```

```python
y = [0,0.1,0.2,0.35,0.5,0.4,0.8,1,1.4,1.6]
make_graph(y)
```

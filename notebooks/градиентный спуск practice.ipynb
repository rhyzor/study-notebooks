```python
# Импортируем необходимые библиотеки 

import sympy as sp
import numpy as np
```

```python
# обозначим переменные символы 

x, y, z = sp.symbols('x, y, z')
```

```python
# напишем функцию MSE для нашей задачи

MSE = 1/6 * (
    (5*x + 7*y - 5*z + 47)**2 +
    (-2*y + 2*z - 10)**2 +
    (-4*x - 8*y - 7*z - 63)**2 +
    (x + y + 2*z + 1)**2 +
    (2*x - y + 2*z + 4)**2 +      
    (4*x + y + 4*z + 2)**2
)

MSE
```

```python
# вычислим производную MSE по x

MSE_x = sp.diff(MSE, x)
MSE_x
```

```python
# вычислим производную MSE по y

MSE_y = sp.diff(MSE, y)
MSE_y
```

```python
# вычислим производную MSE по z

MSE_z = sp.diff(MSE, z)
MSE_z
```

```python
# обозначим скорость обучения для градиентного спуска

step = 0.01
```

```python
# зададим начальные координаты в 0, 0, 0

point = np.array([0, 0, 0])
point
```

```python
# посчитаем MSE в координатах 0, 0, 0

MSE.subs({x: 0, y: 0, z: 0})
```

```python
# посчитаем градиент создав вектор из производных и подставив
# значения координат в x, y, z

gradient = np.array([
    MSE_x.subs({x: point[0], y: point[1], z: point[2]}),
    MSE_y.subs({x: point[0], y: point[1], z: point[2]}),
    MSE_z.subs({x: point[0], y: point[1], z: point[2]})
])

gradient
```

```python
# посчитаем следующую точку градиентного спуска

next_point = point - step * gradient
next_point
```

```python
# проверим MSE для полученых значений

MSE.subs({
    x: next_point[0], y: next_point[1], z: next_point[2]
})
```

```python
i = 1
while i <= 100:
    point = next_point

    gradient = np.array([
        MSE_x.subs({x: point[0], y: point[1], z: point[2]}),
        MSE_y.subs({x: point[0], y: point[1], z: point[2]}),
        MSE_z.subs({x: point[0], y: point[1], z: point[2]})
    ])

    next_point = point - step * gradient
    
    print(F'Итерация #{i}, MSE = {MSE.subs({x: next_point[0], y: next_point[1], z: next_point[2]})}')
    i += 1
```

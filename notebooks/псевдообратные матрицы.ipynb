```python
import sympy as sp
import numpy as np
```

```python
X = sp.Matrix([
    [5, 7, -5],
    [0, -2, 2],
    [-4, -8, -7],
    [1, 1, 2],
    [2, -1, 2],
    [4, 1, 4]
    
])
```

```python
X
```

```python
b = sp.Matrix([
    -47, 10, 63, -1, -4, -2
])
```

```python
b
```

```python
X.transpose()
```

```python
X.T
```

```python
mul = X.T * X
mul
```

```python
mul.det()
```

```python
(X.T * X) ** (-1)
```

```python
(X.T * X).inv()
```

```python
(X.T * X) ** (-1) == (X.T * X).inv()
```

```python
# находим псевдообратную матрицу по формуле

(X.T * X).inv() * X.T
```

```python
# находим вектор весов умножая на b 

weights = (X.T * X).inv() * X.T * b

weights
```

```python
x, y, z = sp.symbols('x, y, z')
```

```python
mse = 1/6 * (
    (5*x + 7*y - 5*z + 47)**2 + 
    (-2*y + 2*z - 10)**2 +
    (-4*x - 8*y - 7*z - 63)**2 +
    (x + y + 2*z + 1)**2 +
    (2*x - y + 2*z + 4)**2 +
    (4*x + y + 4*z + 2)**2
)
```

```python
mse.subs({x: weights[0], y: weights[1], z: weights[2]})
```

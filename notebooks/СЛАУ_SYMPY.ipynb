```python
import sympy as sp
import numpy
```

```python
a = sp.Matrix([
    [2, 5, 11],
    [7, -3, -23]
])

x, y = sp.symbols('x, y')
```

```python
sp.linsolve(a, [x,y])
```

```python
b = sp.Matrix([
    [-3, 1, -2],
    [3, 5, 8]
])
```

```python
sp.linsolve(b, [x,y])
```

```python
c = sp.Matrix([
    [2, 3, 12],
    [3, -1, 7]
])
```

```python
sp.linsolve(c, [x, y])
```

```python
d = sp.Matrix([
    [1, 1, 2, -1],
    [2, -1, 2, -4],
    [4, 1, 4, -2]
])
```

```python
z = sp.Symbol('z')
sp.linsolve(d, [x,y,z])
```

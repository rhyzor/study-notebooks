```python
import sympy as sp
```

```python
a0,a1,a2 = sp.symbols('a0,a1,a2')
```

```python
data = [
    (200,2,200),
    (300,1,450),
    (600,3,550)
]
```

```python
# Модель: y_hat = a0 + a1*x1 + a2*x2
```

```python
MSE = (1/3) * ( ((a0+a1*200+a2*2-200)**2) +
                ((a0+a1*300+a2*1-450)**2) +
                ((a0+a1*600+a2*3-550)**2) 
              )

MSE
```

```python
diff_MSE_a0 = sp.diff(MSE, a0)
diff_MSE_a1 = sp.diff(MSE, a1)
diff_MSE_a2 = sp.diff(MSE, a2)
```

```python
diff_MSE_a0
```

```python
diff_MSE_a1
```

```python
diff_MSE_a2
```

```python
solution = sp.nonlinsolve([diff_MSE_a0, diff_MSE_a1, diff_MSE_a2], [a0, a1, a2])

print("Оптимальные коэффициенты:")
print(solution)
```

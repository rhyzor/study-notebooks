```python
import sympy as sp

A = sp.Matrix([[sp.Rational(3,10), -2, sp.Rational(1,2)]])
B = sp.Matrix([[7, 13], [-4, 5], [5, 6]])

C = A * B
print(C)  # Matrix([[63/5, -31/10]])
```

```python
import numpy as np

A = np.array([[3, 4], [10, 2]])
B = np.array([[4, 8], [2, 4]])

C = A @ B
print(C)  # [[20 40]
          #  [44 88]]
```

```python
A = np.array([[3, 4], [10, 2], [10, 2]])
B = np.array([[4, 8], [2, 4], [10, 2]])

C = A * B
print(C)
```
